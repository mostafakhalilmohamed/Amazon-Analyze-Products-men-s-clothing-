## Extract data from 300 pages 48 result per page
> i put every record in it try and excepterror instead of put all recoed between only one try and except \n cause if i put between one try and except any record have null value will be all the row null but if every record in it try and ecept so if the record is null that will  not affect on other record

> put the result in tuple instead of list to avoid duplicated(i think that)

```python
list1 = []
for page in range(1,301):
    url = 'https://www.amazon.com/s?i=fashion-mens-intl-ship&bbn=16225019011&rh=n%3A7141123011%2Cn%3A16225019011%2Cn%3A1040658&page={}&pf_rd_i=16225019011&pf_rd_m=ATVPDKIKX0DER&pf_rd_p=554625a3-8de1-4fdc-8877-99874d353388&pf_rd_r=B6C2EE7GM8PG44K8XYSX&pf_rd_s=merchandised-search-4&pf_rd_t=101&qid=1620942808&ref=sr_pg_2'
    driver.get(url.format(page))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    cards = soup.find_all('div',{'class':'sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col sg-col-4-of-20'})
    for card in cards:
        product = card.find('span',{'class':'a-size-base-plus a-color-base a-text-normal'}).text
        try:
            rating = card.find('span','a-icon-alt').text.replace('out of 5 stars','')
        except AttributeError:
            rating = ('')
        try:    
            no_review = card.find('span','a-size-base').text
        except AttributeError:
            no_review = ('')
        try:    
            price = card.find('span','a-offscreen').text.replace('$','')
        except AttributeError:
            price = ('')
        try:    
            get_img = card.find('img',{'class':'s-image'})
            image = get_img.get('src')
        except AttributeError:
            image = ('')
        try:    
            get_url = card.find('a',{'class':'a-link-normal s-no-outline'})
            product_url ='https://www.amazon.com/' + get_url.get('href')
        except AttributeError:
            product_url = ('')
            

        result = (product,rating,no_review,price,image,product_url)
        list1.append(result)
```

```python
len(list1)
```

## The process of scrapping 300 pages with 48 result per page too around 30 min

```python
df = pd.DataFrame(list1, columns=['product','rating','no_review','price','image','product_url'])
```
